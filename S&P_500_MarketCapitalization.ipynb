```python
def load_sp_data_from_wikipedia():

  # There are 2 tables on the Wikipedia page
  # we want the first table

  payload=pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
  first_table = payload[0]
  second_table = payload[1]

  df = first_table
  symbols = df['Symbol'].values.tolist()
  print(symbols[:15])
  return df, symbols

df, symbols=load_sp_data_from_wikipedia() 
```

## Get Market Cap

```python
def get_marketcap():
  #marketcap_dict={}
  market_info_df = pd.DataFrame(columns=['Symbol', 'MarketCap', 'ForwardPE', 'TrailingPE'])

  for symbol in df['Symbol']:
    sys=symbol.replace('.', '-')  #BRk.B --> BRK-B
    market_cap=0.0
    forward_pe = 0.0
    trailing_pe=0.0
    info = yf.Ticker(sys).info
    try:
      market_cap = info['marketCap']
    except KeyError:
      market_cap = 0.0
      print(symbol, 'Error marketCap')
    try:
      forward_pe = info['forwardPE']
    except KeyError:
        forward_pe = 0.0 
        print(symbol, 'Error forwardPE')
    try:
      trailing_pe = info['trailingPE']
    except KeyError:
      trailing_pe = 0.0
      print(symbol, 'Error trailingPE')
    market_info_df.loc[len(market_info_df.index)] = [symbol, market_cap, forward_pe, trailing_pe] 
    print(symbol, 'Market Cap: ', market_cap, ' - ', forward_pe, ' -: ', trailing_pe)

  return market_info_df

market_info_df = get_marketcap()
```
